```python
data = fits.getheader('./metric/metric_10d.fts')
```

```python
d1 = get_sft_as_arrays("./data/real_data/H-7338_H1_1800SFT_O4RUN+R1+CGDSCALIBSTRAINCLEANGATEDG02+WTKEY5_NBF0099Hz0W0003Hz0-1368980712-20472783.sft")
h1_time = d1[1]['H1']

d1 = get_sft_as_arrays("./data/real_data/L-7622_L1_1800SFT_O4RUN+R1+CGDSCALIBSTRAINCLEANGATEDG02+WTKEY5_NBF0099Hz0W0003Hz0-1368976350-20451259.sft")
l1_time = d1[1]['L1']
```
